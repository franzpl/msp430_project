# MSP430 PWM

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
clk = 500000 # Hz
freq = 130 # Hz
pulse_width = 0.00006 # seconds
```

```python
t = np.rint(clk / freq).astype(int) # computing the relation between the desired frequency and the clock frequency
```

```python
CCR0 = t - 1 # # set CCR0 to the desired frequency
CCR1 = np.rint(pulse_width * clk).astype(int) # set CCR1 to the desired pulse width
```

## Resolution

The microcontroller works with integer values. Therefore, the minimum value for CCR1 is 1. CCR0 is fixed by the clock frequency and the desired PWM frequency. Consequently, the smallest reachable duty cycle is: 

```python
duty_cycle_resolution = 1 / t
```

```python
duty_cycle_resolution 
```

Thus, the resolution for the pulse width is: 

```python
pulse_width_resolution = 1 / clk
```

```python
pulse_width_resolution
```

## All adjustable duty cycles / pulse widths

```python
plt.plot( np.arange(t), np.arange(t) / t * 100)
plt.ylabel('Duty Cycle / %')
plt.xlabel('CCR1')
plt.title('All adjustable duty cycles')
plt.grid()
plt.show()
```

```python
plt.plot( np.arange(t), np.arange(t) * 1 / clk * 1000000)
plt.ylabel('Pulse Width / us')
plt.title('All adjustable pulse widths')
plt.xlabel('CCR1')
plt.grid()
plt.show()
```

## Conclusion

It will be sufficient that the PWM pulse width is adjustable in steps of 2 microseconds. Therefore, the clock can run with 500 kHz.

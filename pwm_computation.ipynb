The duty cycle is defined as the ratio between the pulse width and the period of a rectangular waveform. A duty cycle can be expressed as:

$D = \frac{PW}{T} \cdot 100 \% $

$D = PW \cdot f \cdot 100\% $

## Stimulation parameters

```python
amplitude = 100e-6 # uA
f = 130 # Hz (T = 1 / 130 = 7.6ms)
pulse_width = 60e-6 # us
```

```python
D = pulse_width * 130
```

The DutyCycle is:

```python
D
```

## Energy Consumption

The average of the consumed current is:

```python
D * amplitude
```

In that case, we will consume 1 uA in average!
